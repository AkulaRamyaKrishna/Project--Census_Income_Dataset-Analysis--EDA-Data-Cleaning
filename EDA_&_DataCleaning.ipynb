# EDA-6 : Identify the Outliers

```python
# Definition: Outliers are extreme values that deviates significantly form rest of the data

# Interview Question : Is categorical column can have outliers ?
# Answer             : Yes there can be outlier in categorical column also. there may be one class which is repeated so many times who's frequency is 
#                      very high or there might be a class who's frequency is very low


### Identifing Outliers there are two Methods:
# 1. IQR (interquantile range)  - It tells the exact value of outlier
# 2. Box  plot
```

```python
Q1 = df.age.quantile(.25)
```

```python
Q3 = df.age.quantile(.75)
```

```python
Q1,Q3
```

```python
IQR = Q3 - Q1
IQR
 
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python

lower_bound,upper_bound
```

```python
df[df.age > 78]
```

# Option-3: If your data is Normally Distibuted
#### Z-score to identify and trim the outliers
#### Z-Score: How much std data point is far away from the mean
#### Z-Score = (data - mean) / std

```python
df11 = df.copy()
```

```python
# Calculate the z score of all data points
df11['z_score'] = (df.age - df.age.mean()) / df.age.std()
```

```python
df11.head()
```

```python
df11[(df11.z_score > 3) | (df11.z_score < -3)].shape
```

```python
df11 = df11[(df11.z_score <= 3) & (df11.z_score >= -3)]
```

```python
df11.shape
```
